I create 5 past-value variables for each variable in my dataframe. To add them to my dataframe, I get rid of my first 5 columns, for which 5 past quarter values are not given. Also, maximum, minimum and average values for 3 and 5 past quarters of all variables are added to the dataframe.

Lastly, I create a standardized version of my dataframe.
